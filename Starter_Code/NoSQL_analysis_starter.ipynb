### 1. Which establishments have a hygiene score equal to 20?

```python
establishments.find_one()
```

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': 20}

# Use count_documents to display the number of documents in the result
hygiene_score = establishments.count_documents(query)
print(f'The number of documents with a Hygiene score of 20 is: {hygiene_score}')
print('-----------------------')

# Display the first document in the results using pprint
results = list(establishments.find(query))
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
results_df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print(f'The total number of rows in the dataframe is: {len(results_df)}')
print('------------------------------------')

# Display the first 10 rows of the DataFrame
results_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'LocalAuthorityName': {'$regex': 'London'},
            'RatingValue': {'$gte': 4}}

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f'The number of documents with London as the Local Authority and a RatingValue greater than or equal to 4 is: {count}')
print('--------------------------------------------------------------------------------------------------------')

# Display the first document in the results using pprint
count_results = list(establishments.find(query))
pprint(count_results[0])
```

```python
# Convert the result to a Pandas DataFrame
count_results_df = pd.DataFrame(count_results)

# Display the number of rows in the DataFrame
print(f'The total number of rows in the dataframe is: {len(count_results_df)}')

# Display the first 10 rows of the DataFrame
count_results_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.490142
longitude = 0.08384

query = {'RatingValue': 5, 
         'geocode.latitude': {'$gte': latitude - degree_search,'$lte': latitude + degree_search},
         'geocode.longitude': {'$gte': longitude - degree_search, '$lte': longitude + degree_search}}
# sort =  {'scores.Hygiene':1}

results = list(establishments.find(query).sort('scores.Hygiene'))

# Print the results
for x in results:
    print(x)
```

```python
# Convert result to Pandas DataFrame
nearby_restaurants_df = pd.DataFrame(results)
nearby_restaurants_df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
match_stage = {'$match': {'scores.Hygiene': 0}}

# 2. Groups the matches by Local Authority
group_stage = {'$group': {'_id': {'LocalAuthorityName': '$LocalAuthorityName'},
                        'count': {'$sum': 1}
                        }}

# 3. Sorts the matches from highest to lowest
sort_stage = {'$sort':{'count': -1}}

pipeline = [match_stage, group_stage, sort_stage]

pipeline_results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print("Number of documents in result: ", len(pipeline_results))
print('--------------------------------')

# Print the first 10 results
for row in pipeline_results[:10]:
    print(row)
```

```python
# Convert the result to a Pandas DataFrame
pipeline_results_df = pd.DataFrame(pipeline_results)

# Display the number of rows in the DataFrame
print(f'The total number of rows in the dataframe is: {len(pipeline_results_df)}')
print('--------------------------------------------------------------')

# Display the first 10 rows of the DataFrame
pipeline_results_df.head(10)
```
